```python
df2017 = pd.read_csv('data/test2017.csv',encoding='CP949')
df2017.head()
```

```python
df2017.columns=["대여일자","대여시간","대여소ID","대여소명","대여구분코드","성별","연령대코드","이용건수","운동량","탄소량","이동거리","이동시간"]
```

```python
df2017=df2017.replace(r'[,\"\']','', regex=True).replace(r'\s*([^\s]+)\s*', r'\1', regex=True)
```

```python
rentals['대여소ID']=rentals['대여소ID'].astype('float')
rentals['대여소ID']
```

```python
#df2017+rentals 데이터 프레임 합치기
#얼마만큼의 데이터가 손실되는지
before2017 = df2017.shape[0]
after2017 = pd.merge(df2017, rentals[['대여소ID', '대여소_구']], on='대여소ID', how='inner').shape[0]
print("매핑하기 이전:", before2017)
print("매핑하기 이후:", after2017)
```

```python
#손실 1프로 정도는 ㄱㅊ
(before2017 - after2017) / before2017 * 100
```

```python
before2018 = df2018.shape[0]
after2018 = pd.merge(df2018, rentals[['대여소ID', '대여소_구']], on='대여소ID', how='inner').shape[0]
print("매핑하기 이전:", before2018)
print("매핑하기 이후:", after2018)
```

```python
#손실 0.6 ㄱㅊ
(before2018 - after2018) / before2018 * 100
```
