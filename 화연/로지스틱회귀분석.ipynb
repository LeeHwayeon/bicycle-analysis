```python
#사망자수와 중상자수 더한 값을 새로운 데이터프레임 열에 저장
attacker['severity'] = attacker['사망자수'] + attacker['중상자수']
victim['severity'] = victim['사망자수'] + victim['중상자수']

#사망자나 중상도가 있을 경우 심각도 높음1 낮음0
def function(x):
    if x >=1:
        return '1'
    else :
        return '0'
attacker['severity']=attacker["severity"].apply(function)
victim['severity']=victim["severity"].apply(function)
```

```python
attacker['sex'] = attacker['가해운전자 성별']
victim['sex'] = victim['피해운전자 성별']

#남1 여0
def function(x):
    if x =="남":
        return '1'
    else :
        return '0'
attacker['sex']=attacker["sex"].apply(function)
victim['sex']=victim["sex"].apply(function)
```

```python
#나이값에 포함되어있는 "세" 제거
attacker['age']=attacker['가해운전자 연령'].map(lambda x:x.replace('세',''))
attacker['age']
```

```python
#나이값에 포함되어있는 "세" 제거
victim['피해운전자 연령']=victim['피해운전자 연령'].astype(str)
victim['age']=victim['피해운전자 연령'].map(lambda x:x.replace('세',''))
victim['age']
```

```python
attacker['law_violation'] = attacker['법규위반']
victim['law_violation'] = victim['법규위반']

#교통법규 위반 유1 무0
def function(x):
    if x in ("교차로운행방법위반","보행자보호의무위반","불법유턴","신호위반","중앙선침범","직진우회전진행방해","차로위반") :
        return '1'
    else :
        return '0'
attacker['law_violation']=attacker["law_violation"].apply(function)
victim['law_violation']=victim["law_violation"].apply(function)
```

```python
#도로종류 교차로1 나머지0
attacker['road_type'] = attacker['도로형태']
victim['road_type'] = victim['도로형태']

def function(x):
    if x in ("교차로 - 교차로부근","교차로 - 교차로안","교차로 - 교차로횡단보도내") :
        return '1'
    else :
        return '0'
attacker['road_type']=attacker["road_type"].apply(function)
victim['road_type']=victim["road_type"].apply(function)
```

```python
#맑음0 나머지1
attacker['weather'] = attacker['기상상태']
victim['weather'] = victim['기상상태']

def function(x):
    if x =='맑음' :
        return '0'
    else :
        return '1'
attacker['weather']=attacker["weather"].apply(function)
victim['weather']=victim["weather"].apply(function)
```

```python
#발생일시에서 시간만 뽑고 "시" 문자열 제거
victim['time']=victim['발생일시'].str[-3:]
victim['time']=victim['time'].map(lambda x:x.replace('시',''))
victim['time']
```

```python
#발생일시에서 시간만 뽑고 "시" 문자열 제거
attacker['time']=attacker['발생일시'].str[-3:]
attacker['time']=attacker['time'].map(lambda x:x.replace('시',''))
attacker['time']
```

```python
#시간
#주간(오전7시~오후6시)-0  야간(오후7시~오전6시)-1
def function(x):
    if x in ("19","20","21","22","23","00","01","02","03","04","05","06") :
        return '1'
    else :
        return '0'
attacker['time']=attacker["time"].apply(function)
victim['time']=victim["time"].apply(function)
```

```python
attacker['roadbed']=attacker['노면상태']
victim['roadbed']=victim['노면상태']

#노면상태 건조0 기타1
def function(x):
    if x in ("포장 - 건조","비포장 - 건조") :
        return '0'
    else :
        return '1'
attacker['roadbed']=attacker["roadbed"].apply(function)
victim['roadbed']=victim["roadbed"].apply(function)
```

### 로지스틱 회귀분석
모델의 회귀계수와 오즈비를 구해 독립변수가 분류 결정에 미치는 영향의 정도를 알아본다

```python
import statsmodels.api as sm
#가해자
#로지스틱 회귀분석
logit = sm.Logit(accident_attacker[['severity']],x_attacker)
attacker_result = logit.fit()
```

유의확률 p>|z|가 0.05보다 작으면 유의수준 95%에서 종속변수에 영향을 주는 유의한 변수임
## 유의수준 95%에서 종속변수에 영향을 주는 변수 : 성별(남), 나이(?), 도로형태(교차로)

```python
attacker_result.summary2()
```

### coef(편회귀계수)
- 한 독립변수 외의 다른 독립변수들이 모두 일정한 상태에서 한 독립변수가 1단위 변할 때 종속변수가 얼마나 변하는 가를 알려줌  
- coef를 통해 종속변수에 미치는 영향의 방향을 파악할 수 있다.  
- coef가 양수이면 그 컬럼의 값이 커질수록 종속변수가 TRUE일 확률이 높아진다는 뜻  
- 반대로 음수이면 그 컬럼의 값이 커질수록 종속변수가 FALSE일 확률이 높아진다는 뜻  
## 자전거 가해운전자 데이터 : 법규위반(o),날씨(맑음x),시간(오후7시~오전6시)

```python
np.exp(attacker_result.params)
```

### 오즈비 : 오즈의 비율
- Odds란 성공/실패와 같이 상호 배타적이며 전체를 이루고 있는 것들의 비율을 의미  
- 오즈비가 1을 기준으로 큰지 작은지를 파악하여 종속변수에 미치는 영향의 방향을 파악할 수 있다.  
- 아무런 관계가 없을 때 오즈비는 1이다. 1에서 멀리 떨어질수록 종속변수와의 관계가 강하다는 뜻이다. 즉, 종속변수 여부에 큰 영향을 준다는 뜻이다.  
## 자전가 가해운전자 데이터: 성별(남), 도로형태(교차로)가 영향을 미침
