```python
from pandas import DataFrame as df
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import missingno as msno
import warnings
warnings.filterwarnings('ignore')

mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams["font.family"] = 'Malgun Gothic'

import json
import folium
```

```python
#따릉이 이용현황 : 2017년과 2018년 합친 데이터 불러오기
bicycle = pd.read_csv('data/bicycle.csv', parse_dates=["대여일자"],encoding='CP949')
```

```python
bicycle
```

```python
df=df({'대여일자':bicycle['대여일자'], '대여시간':bicycle['대여시간'],'이용건수':bicycle['이용건수']})
```

```python
df
```

```python
#대여시간의 데이터의 형태(?)가 일정하지 않음
#0~9는 00~09와 합쳐져야 함
def function(x):
    if x =='0':
        return '00'
    elif x=='1':
        return '01'
    elif x=='2':
        return '02'
    elif x=='3':
        return '03'
    elif x=='4':
        return '04'
    elif x=='5':
        return '05'
    elif x=='6':
        return '06'
    elif x=='7':
        return '07'
    elif x=='8':
        return '08'
    elif x=='9':
        return '09'
    else :
        return x
df['대여시간']=df['대여시간'].apply(function)
df
```

```python
df=df.dropna()
df
```

```python
df['대여시간']=df['대여시간'].astype(int)
df['이용건수']=df['이용건수'].astype(int)
```

```python
df.info()
```

```python
df['월']=df['대여일자'].dt.month
df['요일']=df['대여일자'].dt.dayofweek
```

```python
accident_month=pd.read_csv('data/accident_month.csv',encoding='euc-kr')
accident_dayofweek=pd.read_csv('data/accident_dayofweek.csv',encoding='euc-kr')
accident_time=pd.read_csv('data/accident_time.csv',encoding='euc-kr')
```

```python
accident_month
```

```python
accident_dayofweek
```

```python
accident_time
```

```python
accident_month = accident_month.groupby(['월']).agg('sum')
accident_dayofweek = accident_dayofweek.groupby(['요일']).agg('sum')
accident_time = accident_time.groupby(['시간']).agg('sum')
```

```python
accident_month.info()
accident_dayofweek.info()
accident_time.info()
```

```python
#자전거 이용현황 시간별 데이터와 자전거 교통사고 시간별 데이터 기준이 달라서 같게 만들어야함
#자전거 교통사고를 기준으로 데이터 만들기로 함
time=df.groupby(["대여시간"])[["이용건수"]].count()
time
```

```python
#자전거 이용시간 데이터를 교통사고 데이터처럼 합치는 작업
time1=time.iloc[0]+time.iloc[1]
time2=time.iloc[2]+time.iloc[3]
time3=time.iloc[4]+time.iloc[5]
time4=time.iloc[6]+time.iloc[7]
time5=time.iloc[8]+time.iloc[9]
time6=time.iloc[10]+time.iloc[11]
time7=time.iloc[12]+time.iloc[13]
time8=time.iloc[14]+time.iloc[15]
time9=time.iloc[16]+time.iloc[17]
time10=time.iloc[18]+time.iloc[19]
time11=time.iloc[20]+time.iloc[21]
time12=time.iloc[22]+time.iloc[23]
```

```python
time1
```

```python
from pandas import DataFrame as df

#새로운 시간대별 자전거 이용건수 데이터 프레임 생성
data=["00시~02시","02시~04시","04시~06시","06시~08시","08시~10시","10시~12시","12시~14시","14시~16시","16시~18시","18시~20시","20시~22시","22시~24시"]
bicycle_time=df([time1,time2,time3,time4,time5,time6,time7,time8,time9,time10,time11,time12])
bicycle_time['시간']=df(data)
```

```python
bicycle_time
```

```python
bicycle_time=bicycle_time.groupby(["시간"])[['이용건수']].agg('sum')
ax=bicycle_time.plot(kind='bar', figsize=(12, 8), rot=0, title="시간별 자전거 대여건수")
for p in ax.patches: 
    left, bottom, width, height = p.get_bbox().bounds 
    ax.annotate("%d "%(height), (left+width/2, height*1.03), ha='center')
plt.box(False)
plt.legend(frameon=False)
plt.show()
```

```python
#시간대별 자전거 이용x교통사고
ac_time=accident_time.groupby(['시간'])[['사고건수']].agg('sum')
ax=ac_time.plot(kind='bar', figsize=(12, 8), rot=0, title="시간대별 자전거 사고건수")
for p in ax.patches: 
    left, bottom, width, height = p.get_bbox().bounds 
    ax.annotate("%d"%(height), (left+width/2, height*1.03), ha='center')
plt.box(False)
plt.legend(frameon=False)

plt.show()
```

```python
#시간대별 자전거 이용x교통사고
ac_time=accident_time.groupby(['시간'])[['사고건수']].agg('sum')
ax=ac_time.plot(kind='bar', figsize=(12, 8), rot=0, title="시간대별 자전거 대여건수와 사고건수")
for p in ax.patches: 
    left, bottom, width, height = p.get_bbox().bounds 
    ax.annotate("%d"%(height), (left+width/2, height*1.03), ha='center')
plt.box(False)
plt.legend(frameon=False)

plt.twinx() # bar 그래프와 plot 그래프 합치기
plt.plot(bicycle_time,'r.-')
plt.show()
```
