```python
import pandas as pd
%matplotlib inline 
 
## matplotlib의 pyplot import
import matplotlib
import matplotlib.pyplot as plt

import matplotlib.font_manager as fm
```

```python
accident2018=pd.read_csv('data/2018교통사고_요일.csv',encoding='euc-kr')
accident2017=pd.read_csv('data/2017교통사고_요일.csv',encoding='euc-kr')
```

```python
accident2018
```

```python
accident2017
```

```python
#nan 제거
accident2017=accident2017.dropna(how='all',axis=0)
accident2017=accident2017.dropna(how='all',axis=1)
accident2017
```

```python
#nan 제거
accident2018=accident2018.dropna(how='all',axis=0)
accident2018=accident2018.dropna(how='all',axis=1)
accident2018
```

```python
accident2018.info()
```

```python
accident2018.isnull().sum()
```

```python
accident2018['사고건수']=accident2018['사고건수'].astype(int)
accident2018['사망자수']=accident2018['사망자수'].astype(int)
accident2018['부상자수']=accident2018['부상자수'].astype(int)
accident2018.info()
accident2017['사고건수']=accident2017['사고건수'].astype(int)
accident2017['사망자수']=accident2017['사망자수'].astype(int)
accident2017['부상자수']=accident2017['부상자수'].astype(int)
accident2017.info()
```

```python
accident2018.plot()
```

```python
accident2017.plot()
```

```python
font_location='NanumGothic.ttf'
font_naem =fm.FontProperties(fname=font_location).get_name()
```

```python
accident2017.plot(kind='bar', figsize=(12,5),rot=0,title="요일별 교통사고")
accident2017.index=accident2017['요일']
plt.rc('font', family='NanumGothic')
plt.box(False)
plt.tight_layout()
plt.show()
```
