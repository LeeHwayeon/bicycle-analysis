```python
import pandas as pd
%matplotlib inline 
 
## matplotlib의 pyplot import
import matplotlib
import matplotlib.pyplot as plt

import matplotlib.font_manager as fm
```

```python
accident2018=pd.read_csv('data/2018교통사고_월.csv',encoding='euc-kr')
accident2017=pd.read_csv('data/2017교통사고_월.csv',encoding='euc-kr')
```

```python
accident2018=accident2018.dropna(how='all',axis=1)
accident2017=accident2017.dropna(how='all',axis=1)
```

```python
font_location='NanumGothic.ttf'
font_naem =fm.FontProperties(fname=font_location).get_name()
```

```python
accident2017.plot(kind='bar', figsize=(12,5),rot=0,title="월별 교통사고")
accident2017.index=accident2017['월']
plt.rc('font', family='NanumGothic')
plt.box(False)
plt.tight_layout()
plt.show()
```
